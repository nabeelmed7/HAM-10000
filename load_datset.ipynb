```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from glob import glob
from PIL import Image
```

```python
df = pd.read_csv('HAM10000_metadata')
```

```python
df.head(5)
```

```python
path = {os.path.splitext(os.path.basename(x))[0]: x
                     for x in glob(os.path.join('Dataset/', '*', '*.jpg'))}
```

```python
path
```

```python
df['path'] = df['image_id'].map(path.get)
```

```python
df['image'] = df['path'].map(lambda x: np.asarray(Image.open(x).resize((32,32))))
```
